```python
import pandas as pd
```

```python
data_trade_1 = pd.read_csv(
    "data_raw/TradeData_12_2_2025_21_49_21 - FoodRelated.csv",
    encoding='ISO-8859-1'
)
data_trade_2 = pd.read_csv(
    "data_raw/TradeData_12_2_2025_21_54_20 - ProductionRelated.csv",
    encoding='ISO-8859-1'
)
```

```python
data_trade_1.head()
```

```python
data_trade_2.head()
```

```python
data_trade = pd.concat([data_trade_1,data_trade_2], axis=0, ignore_index=True)
```

```python
print(data_trade_1.shape)
print(data_trade_2.shape)
print(data_trade.shape)
```

```python
data_trade.to_csv("data_processed/data_trade.csv", index=False)
```
