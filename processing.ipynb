```python
import pandas as pd
import numpy as np
```

```python
data_trade_2025 = pd.read_csv(
    "data_raw/TradeData_12_15_2025_18_6_30.csv",
    encoding='ISO-8859-1'
)
data_trade_2024 = pd.read_csv(
    "data_raw/TradeData_12_15_2025_17_59_40.csv",
    encoding='ISO-8859-1'
)

data_trade_2023 = pd.read_csv(
    "data_raw/TradeData_12_15_2025_18_29_57.csv",
    encoding='ISO-8859-1'
)
data_trade_2022 = pd.read_csv(
    "data_raw/TradeData_12_15_2025_18_31_33.csv",
    encoding='ISO-8859-1'
)
data_trade_2021 = pd.read_csv(
    "data_raw/TradeData_12_15_2025_18_34_44.csv",
    encoding='ISO-8859-1'
)
data_hcode = pd.read_csv(
    "data_raw/harmonized-system.csv",
    encoding='ISO-8859-1'
)
data_iso = pd.read_csv(
    "data_raw/countries_data.csv"
)
```

```python
data_trade = pd.concat([data_trade_2025, data_trade_2024, data_trade_2023, data_trade_2022, data_trade_2021], axis=0, ignore_index=True)
```

```python
data_trade.isna().sum()
```

```python
data_trade.head(n=1000)
```

```python
data_date = pd.DataFrame(data_trade["freqCode"].drop_duplicates())
```

```python
data_date.head()
```

```python
data_date['freqCode'] = data_date['freqCode'].astype(str)
data_date["Year"] = data_date['freqCode'].str.slice(start=0, stop=4)
data_date["Month_Number"] = data_date['freqCode'].str.slice(start=4, stop=6)
data_date["date"] = data_date["Year"] + "-" + data_date["Month_Number"] + "-01"
```

```python
month_map = {
    "01": "January",
    "02": "February",
    "03": "March",
    "04": "April",
    "05": "May",
    "06": "June",
    "07": "July",
    "08": "August",
    "09": "September",
    "10": "October",
    "11": "November",
    "12": "December"
}

data_date["Month"] = data_date["Month_Number"].map(month_map)

data_date.head()
```

```python
data_date.set_index("freqCode")
```

```python
data_date.to_csv("data_processed/dim_date.csv", index=False)
```

```python
data_trade.drop(["refPeriodId", "refYear", "refMonth", "period"], axis=1, inplace=True)
```

```python
data_trade.drop(["reporterISO"], axis=1, inplace=True)
```

```python
data_flow = data_trade[["reporterDesc", "flowCode"]].drop_duplicates(subset=["reporterDesc"])
data_flow.head()
```

```python
flow_map_group = {
    "DX" : "Export",
    "FM" : "Import",
    "RM" : "Import",
    "RX" : "Export",
    "MIP" : "Import",
    "XIP" : "Export",
    "XOP" : "Export",
    "MOP" : "Import"
}
data_flow["flowGroup"] = data_flow["reporterDesc"].map(flow_map_group)
```

```python
flow_map_type = {
    "Domestic Export" : "Standard",
    "Foreign Import" : "Standard",
    "Re-import" : "Repeated",
    "Re-export" : "Repeated",
    "Import" : "Total",
    "Export" : "Total",
    "Import of goods for inward processing": "Processing",
    "Export of goods after inward processing": "Processing",
    "Export of goods for outward processing": "Processing",
    "Import of goods after outward processing": "Processing"
}
data_flow["flowType"] = data_flow["flowCode"].map(flow_map_type)
```

```python
data_trade.drop(["typeCode"], inplace=True, axis=1)
```

```python
data_trade.drop(["partnerCode", "partnerISO", "partnerDesc", "partner2Code", "partner2ISO", "partner2Desc", "classificationCode", "classificationSearchCode", "isOriginalClassification", "cmdDesc", "aggrLevel"], inplace=True, axis=1)
```

```python
data_trade = pd.merge(data_trade, data_hcode[["description", "hscode"]], left_on="cmdCode", right_on="description", how="left")
```

```python
data_trade.isnull().sum()
```

```python
data_trade["hscode"] = data_trade["hscode"].fillna("24")
```

```python
data_clearance = data_trade[["customsCode", "isLeaf"]].drop_duplicates(subset=["isLeaf"])
```

```python
clearence_end_use_map = {
    "TOTAL CPC" : "No",
    "Outright exportation" : "Yes",
    "Reimportation in the same state" : "No",
    "Clearance for home use" : "Yes"
}
data_clearance["clearenceEndUse"] = data_clearance["customsCode"].map(clearence_end_use_map)
```

```python
clearence_is_repeated_map = {
    "TOTAL CPC" : "No",
    "Outright exportation" : "No",
    "Reimportation in the same state" : "Yes",
    "Clearance for home use" : "No"
}
data_clearance["clearenceIsRepeated"] = data_clearance["customsCode"].map(clearence_is_repeated_map)
```

```python
clearence_is_domestic = {
    "TOTAL CPC" : "No",
    "Outright exportation" : "No",
    "Reimportation in the same state" : "Yes",
    "Clearance for home use" : "No"
}
data_clearance["clearenceIsDomestic"] = data_clearance["customsCode"].map(clearence_is_domestic)
```

```python
data_clearance.to_csv("data_processed/dim_clearence.csv", index=False)
```

```python
data_trade.drop(["customsDesc", "motDesc", "qtyUnitCode", "qtyUnitAbbr", "qty", "isQtyEstimated", "altQtyUnitCode", "altQtyUnitAbbr", "altQty", "isAltQtyEstimated", "netWgt", "isNetWgtEstimated", "grossWgt", "isGrossWgtEstimated", "cifvalue", "primaryValue", "legacyEstimationFlag", "isReported", "isAggregate", "description"], axis=1, inplace=True)
```

```python
data_mots = data_trade[["mosCode", "motCode"]].drop_duplicates(subset=["mosCode"])
```

```python
mots_speed_map = {
    "Air" : "Fast",
    "Water" : "Slow",
    "Road" : "Medium",
    "Railway" : "Medium",
    "Land" : "Medium"
}
data_mots["motSpeedClass"] = data_mots["motCode"].map(mots_speed_map)
```

```python
mots_cost_map = {
    "Air" : "High",
    "Water" : "Low",
    "Road" : "Medium",
    "Railway" : "Medium",
    "Land" : "Medium"
}
data_mots["motCostClass"] = data_mots["motCode"].map(mots_cost_map)
```

```python
mot_avg_cost_ton_map = {
    "Air" : 1.0,
    "Water" : 0.02,
    "Road" : 0.1,
    "Railway" : 0.04,
    "Land" : 0.07
}
data_mots["motAvgCostTon"] = data_mots["motCode"].map(mot_avg_cost_ton_map)
```

```python
data_mots.to_csv("data_processed/dim_mots.csv", index=False)
```

```python
data_hcode.drop(columns=["section", "level"], inplace=True)
```

```python
data_hcode.to_csv("data_processed/dim_product.csv", index=False)
```

```python
data_iso.drop(["alpha-2", "iso_3166-2", "intermediate-region", "intermediate-region-code", "region-code", "sub-region-code"], axis=1, inplace=True)
```

```python
data_iso.head()
```

```python
data_iso.to_csv("data_processed/dim_country.csv", index=False)
```
